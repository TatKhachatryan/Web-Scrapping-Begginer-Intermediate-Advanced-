```python
import pandas as pd
pd.set_option('display.max_columns', None)
from datetime import datetime
import numpy as np
from bs4 import BeautifulSoup
import requests

import warnings
warnings.filterwarnings('ignore')
```

```python
URL = 'https://www.premierleague.com'

page = requests.get(URL + '/clubs')
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# create a list containing the links for all premier league club stats web pages
clubTags = soup.find_all('a', {'class': 'indexItem'})
clubs = []
for tag in clubTags:
    clubURL = URL + tag['href']
    clubURL = clubURL.replace("overview", "stats")
    clubs.append(clubURL)
    
club_names = ['Bournemouth', 'Arsenal', 'Aston-Villa', 'Brentford', 'Brighton-and-Hove-Albion', 'Chelsea', 'Crystal-Palace', 'Everton', 'Fulham',
 'Leeds-United', 'Leicester-City', 'Liverpool', 'Manchester-City', 'Manchester-United', 'Newcastle-United', 'Nottingham-Forest',
 'Southampton', 'Tottenham-Hotspur', 'West-Ham-United', 'Wolverhampton-Wanderers',]
```

```python
matchesPLayed = []
goals = []
goals_per_game = []
shots = []
shots_on_target = []
shot_accuracy = []
penalties_scored = []
big_chances_created = []
hit_woodwork = []
passes = []
passes_per_match = []
pass_accruacy = []
crosses = []
cross_accuracy = []
clean_sheats = []
goals_conceded = []
goals_conceded_per_month = []
saves = []
tackles = []
tackle_success = []
blocked_shots = []
interceptions = []
clearances = []
headed_clearances = []
areial_battles = []
errors_lead_to_goal = []
own_goals = []
yellow_cards = []
red_cards = []
fouls = []
offsides = []

for club in clubs:
    clubPage = requests.get(club)
    clubSoup = BeautifulSoup(clubPage.content, 'html.parser')

    # matchesPLayed values
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statmatches_played'}))):
            matchesPLayed.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statmatches_played'})[i].text.strip())
    except:
        matchesPLayed.append(None)

    # goals
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statgoals'}))):
            goals.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statgoals'})[i].text.strip())
    except:
        goals.append(None)

    # goals_per_game
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statgoals_per_game'}))):
            goals_per_game.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statgoals_per_game'})[i].text.strip())
    except:
        goals_per_game.append(None)

    # shots
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stattotal_scoring_att'}))):
            shots.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stattotal_scoring_att'})[i].text.strip())
    except:
        shots.append(None)

    # shots_on_target
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statontarget_scoring_att'}))):
            shots_on_target.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statontarget_scoring_att'})[i].text.strip())
    except:
        shots_on_target.append(None)
    
    # shot_accuracy
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statshot_accuracy'}))):
            shot_accuracy.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statshot_accuracy'})[i].text.strip())
    except:
        shot_accuracy.append(None)

    # penalties_scored
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statatt_pen_goal'}))):
            penalties_scored.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statatt_pen_goal'})[i].text.strip())
    except:
        penalties_scored.append(None)
        
    # big_chances_created
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statbig_chance_created'}))):
            big_chances_created.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statbig_chance_created'})[i].text.strip())            
    except:
        big_chances_created.append(None)
        
 
    # hit_woodwork
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stathit_woodwork'}))):
            hit_woodwork.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stathit_woodwork'})[i].text.strip())
    except:
        hit_woodwork.append(None)
   
    # passes
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stattotal_pass'}))):
            passes.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stattotal_pass'})[i].text.strip())
    except:
        passes.append(None)
 
    # passes_per_match
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stattotal_pass_per_game'}))):
            passes_per_match.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stattotal_pass_per_game'})[i].text.strip())
    except:
        passes_per_match.append(None)
   
    # pass_accruacy
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statpass_accuracy'}))):
            pass_accruacy.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statpass_accuracy'})[i].text.strip())
    except:
        pass_accruacy.append(None)
                  
     # crosses
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stattotal_cross'}))):
            crosses.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stattotal_cross'})[i].text.strip())
    except:
        crosses.append(None)
    
     # cross_accuracy
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statcross_accuracy'}))):
            cross_accuracy.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statcross_accuracy'})[i].text.strip())
    except:
        cross_accuracy.append(None)

     # clean_sheats
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statclean_sheet'}))):
            clean_sheats.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statclean_sheet'})[i].text.strip())
    except:
        clean_sheats.append(None)
              
     # goals_conceded_per_month
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statgoals_conceded_per_game'}))):
            goals_conceded_per_month.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statgoals_conceded_per_game'})[i].text.strip())
    except:
        goals_conceded_per_month.append(None)
    
     # saves
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statsaves'}))):
            saves.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statsaves'})[i].text.strip())
    except:
        saves.append(None)
              
     # tackles
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stattotal_tackle'}))):
            tackles.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stattotal_tackle'})[i].text.strip())
    except:
        tackles.append(None)
        
     # tackle_success
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stattackle_success'}))):
            tackle_success.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stattackle_success'})[i].text.strip())
    except:
        tackle_success.append(None)
    # blocked_shots
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statblocked_scoring_att'}))):
            blocked_shots.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statblocked_scoring_att'})[i].text.strip())
    except:
        blocked_shots.append(None)
     
    # interceptions
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statinterception'}))):
            interceptions.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statinterception'})[i].text.strip())
    except:
        interceptions.append(None)
 
    # clearances
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stattotal_clearance'}))):
            clearances.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stattotal_clearance'})[i].text.strip())
    except:
        clearances.append(None)
        
        
     # headed_clearances
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stateffective_head_clearance'}))):
            headed_clearances.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stateffective_head_clearance'})[i].text.strip())
    except:
        headed_clearances.append(None)
    
     # areial_battles
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stataerial_battle_duel_won'}))):
            areial_battles.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stataerial_battle_duel_won'})[i].text.strip())
    except:
        areial_battles.append(None)
        
     # errors_lead_to_goal
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'staterror_lead_to_goal'}))):
            errors_lead_to_goal.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'staterror_lead_to_goal'})[i].text.strip())
    except:
        errors_lead_to_goal.append(None)
        
     # own_goals
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statown_goals'}))):
            own_goals.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statown_goals'})[i].text.strip())
    except:
        own_goals.append(None)
        
     # yellow_cards
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statyellow_card'}))):
            yellow_cards.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statyellow_card'})[i].text.strip())
    except:
        yellow_cards.append(None)
        
     # red_cards
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statred_card'}))):
            red_cards.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statred_card'})[i].text.strip())
    except:
        red_cards.append(None)
    # fouls
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statattempted_tackle_foul'}))):
            fouls.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'statattempted_tackle_foul'})[i].text.strip())
    except:
        fouls.append(None)
    # offsides
    try:
        for i in range(len(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stattotal_offside'}))):
            offsides.append(clubSoup.find_all('span', {'class': 'allStatContainer', 'class': 'stattotal_offside'})[i].text.strip())
    except:
        offsides.append(None)

        
goals = goals[0::2]
clean_sheats = clean_sheats[0::2]   
```
